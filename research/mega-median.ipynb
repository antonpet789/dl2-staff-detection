```python
import cv2
import pandas as pd
import numpy as np

from tqdm import tqdm

TRN_DIR = "./dl-lab-2-stuff-detection/yolo_dataset/yolo_dataset/train/images"
TST_DIR = "./dl-lab-2-stuff-detection/test_images/test_images"

MTRN = pd.read_csv('./trn_meta.csv', parse_dates=['datetime'])
MTST = pd.read_csv('./tst_meta.csv', parse_dates=['datetime'])
```

```python
def impaths():
    def generator():
        for fn in MTRN['fname']:
            yield f"{TRN_DIR}/{fn}"
        for fn in MTST['fname']:
            yield f"{TST_DIR}/{fn}"
    return tqdm(enumerate(generator()), total=len(MTRN) + len(MTST))
```

```python
# n = 8362
H, W, C = 720, 1280, 3
P = H * W * C  # 2 764 800

hist = np.zeros(shape=(P, 256), dtype=np.uint16)  # ~ 1.32 GB
irange = np.arange(P, dtype=np.uint32)            # ~ 10.5 MB

img = None; processed_num = 0
for i, path in impaths():
    img = cv2.imread(path)  # ~ 2.6 MB
    if img is None: continue
    hist[irange, img.ravel()] += 1   # type: ignore
    processed_num += 1
del img

cum_hist = np.cumsum(hist, axis=1, dtype=np.uint16)
median_threshold = processed_num // 2
median_flat = np.argmax(cum_hist >= median_threshold, axis=1)
median_img = median_flat.reshape(H,W,C).astype(np.uint8)

cv2.imwrite('median_result.jpg', median_img)
```
